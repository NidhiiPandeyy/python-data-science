- selecting rows and columns
    - rows (slicing)
        - 'loc' - gets us the data using named indexes
        - 'iloc' - gets us the data using numeric indexes
    - col (dictionary style)
        - df[single] - requires name of column
        - df[multiple] - requires a list of columns (muliple is list of named cols)


```python
# single row (record)
weather.iloc[0]
```

```python
weather.iloc[-1]
```

```python
# records 25 - 50
weather.iloc[25:50]
```

```python
# records 1 - 15 with first 5 col
weather.iloc[1:16, :5]
```

```python
# records 1 - 15 with first 5 col
weather.iloc[1:16, [0,1,2,3,4]]
```

```python
weather.iloc[:15,-3:]
```

```python
weather.iloc[:15, [-3,-2,-1]]
```

```python
weather['Date']
```

```python
weather.AvgTemp  # same as weather['AvgTemp]
```

```python
cols = ['Date','MaxTemp','MinTemp']
weather[cols]  # multiple cols (dictionary style) -> DataFrame
```
